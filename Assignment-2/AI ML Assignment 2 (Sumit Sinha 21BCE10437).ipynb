3. Perform the Below Visualizations. ● Univariate Analysis ● Bi - Variate Analysis
● Multivariate Analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

Univariate Analysis

```python
sns.histplot(df['living area'], bins=5, kde=True)
plt.xlabel('numbers')
plt.ylabel('Frequency')
plt.title('Histogram of Age')
```

Bi - Variate Analysis

```python
sns.scatterplot(data=df, x='number of bedrooms', y='living area', marker='o', color='green')
plt.xlabel('number of bedrooms')
plt.ylabel('living area')
plt.title('Scatter Plot of number of bedrooms vs. living area')
```

Multivariate Analysis

```python
correlation_matrix = df.corr()
```
